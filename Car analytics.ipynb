# Data cleaning

```python
df.dtypes
```

```python
df.isnull().sum()
```

QUANTITAVE

```python
mean=df['EngineSize'].mean()
df['EngineSize'].replace(np.nan,mean, inplace=True)
mean1=df['Cylinders'].mean()
df['Cylinders'].replace(np.nan,mean1, inplace=True)
mean2=df['Horsepower'].mean()
df['Horsepower'].replace(np.nan,mean2, inplace=True)
mean3=df['MPG_City'].mean()
df['MPG_City'].replace(np.nan,mean3, inplace=True)
mean4=df['MPG_Highway'].mean()
df['MPG_Highway'].replace(np.nan,mean4, inplace=True)
mean5=df['Weight'].mean()
df['Weight'].replace(np.nan,mean5, inplace=True)
mean6=df['Wheelbase'].mean()
df['Wheelbase'].replace(np.nan,mean6, inplace=True)
mean7=df['Length'].mean()
df['Length'].replace(np.nan,mean7, inplace=True)
```

```python
df.isnull().sum()
```

QUALITATIVE

```python
r1=df['Make'].value_counts().idxmax()
df["Make"].replace(np.nan, r1, inplace=True)
r2=df['Model'].value_counts().idxmax()
df['Model'].replace(np.nan,r2,inplace=True)
r3=df['Type'].value_counts().idxmax()
df['Type'].replace(np.nan,r3,inplace=True)
r4=df['Origin'].value_counts().idxmax()
df['Origin'].replace(np.nan,r4,inplace=True)
r5=df['DriveTrain'].value_counts().idxmax()
df['DriveTrain'].replace(np.nan,r5,inplace=True)
r6=df['MSRP'].value_counts().idxmax()
df['MSRP'].replace(np.nan,r6,inplace=True)
r7=df['Invoice'].value_counts().idxmax()
df['Invoice'].replace(np.nan,r7,inplace=True)
```
